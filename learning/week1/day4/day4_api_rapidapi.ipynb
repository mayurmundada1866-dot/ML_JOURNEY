```python
import requests
import pandas as pd
```

```python
url = "https://jsonplaceholder.typicode.com/users"
```

```python
response = requests.get(url)
response.status_code
```

```python
data = response.json()
data
```

```python
df = pd.json_normalize(data)
df.head()
```

```python
df.info()
df.columns
```

```python
df.to_csv("users_from_api.csv", index=False)
```
